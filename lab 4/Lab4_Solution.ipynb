## 🔵 Task #1 – Thresholding (Procedural Steps)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, img_as_float, exposure

print("All libraries imported successfully ✅")
```

```python
# ─────────────────────────────────────────────────────────────────
# TASK #1 – Thresholding with Fixed Threshold Values
# ─────────────────────────────────────────────────────────────────

# Load a sample grayscale image from skimage (using 'moon' as substitute for Parrot.png)
img_parrot = data.moon()

# Threshold values to test
threshold_values = [0, 50, 100, 150, 200]

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

# Show original image
axes[0].imshow(img_parrot, cmap='gray')
axes[0].set_title('Original Image', fontsize=13, fontweight='bold')
axes[0].axis('off')

# Apply thresholding with different threshold values
for i, threshold_value in enumerate(threshold_values):
    ret, thresh = cv2.threshold(
        img_parrot.astype(np.uint8),
        threshold_value, 255,
        cv2.THRESH_BINARY
    )
    axes[i + 1].imshow(thresh, cmap='gray')
    axes[i + 1].set_title(f'Threshold = {threshold_value}', fontsize=12)
    axes[i + 1].axis('off')

plt.suptitle('Task #1 – Fixed Threshold Comparison', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('task1_thresholding.png', dpi=150, bbox_inches='tight')
plt.show()
print("Task #1 – Thresholding completed ✅")
```

## 🔵 Task #2 – Histogram Processing (Procedural Steps)

```python
# ─────────────────────────────────────────────────────────────────
# TASK #2 – Histogram Processing (Procedural Steps)
# Step 2: Load moon image
# ─────────────────────────────────────────────────────────────────

img = data.moon()

# Step 3: Contrast Stretching – rescale to 2nd and 98th percentiles
p2, p98 = np.percentile(img, (2, 98))
img_rescale = exposure.rescale_intensity(img, in_range=(p2, p98))

# Step 4: Display original and contrast-stretched images with histograms
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# --- Original Image ---
axes[0, 0].imshow(img, cmap='gray')
axes[0, 0].set_title('Original Image', fontsize=13, fontweight='bold')
axes[0, 0].axis('off')

axes[1, 0].hist(img.flat, bins=256, range=(0, 255), color='steelblue', edgecolor='none')
axes[1, 0].set_xlabel('Gray Level Values')
axes[1, 0].set_ylabel('Number of Pixels')
axes[1, 0].set_title('Histogram – Original', fontsize=12)

# --- Contrast Stretched Image ---
axes[0, 1].imshow(img_rescale, cmap='gray')
axes[0, 1].set_title('Contrast Stretched (2nd–98th %)', fontsize=13, fontweight='bold')
axes[0, 1].axis('off')

axes[1, 1].hist(img_rescale.flat, bins=256, range=(0, 255), color='darkorange', edgecolor='none')
axes[1, 1].set_xlabel('Gray Level Values')
axes[1, 1].set_ylabel('Number of Pixels')
axes[1, 1].set_title('Histogram – Contrast Stretched', fontsize=12)

plt.suptitle('Task #2 – Histogram Processing (Procedural)', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('task2_histogram_processing.png', dpi=150, bbox_inches='tight')
plt.show()
print("Task #2 – Histogram Processing (Procedural) completed ✅")
```

## 🟢 Assessment Task #1 – Contrast Stretching (3rd–80th Percentiles)

```python
# ─────────────────────────────────────────────────────────────────
# ASSESSMENT TASK #1
# Rescale intensity to include all intensities within 3rd and 80th percentiles
# ─────────────────────────────────────────────────────────────────

img = data.moon()

# Calculate 3rd and 80th percentiles
p3, p80 = np.percentile(img, (3, 80))
print(f"3rd Percentile value:  {p3:.2f}")
print(f"80th Percentile value: {p80:.2f}")

# Rescale intensity
img_rescale_3_80 = exposure.rescale_intensity(img, in_range=(p3, p80))

# Plot image and histogram
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# --- Original Image ---
axes[0, 0].imshow(img, cmap='gray')
axes[0, 0].set_title('Original Moon Image', fontsize=13, fontweight='bold')
axes[0, 0].axis('off')

axes[1, 0].hist(img.flat, bins=256, range=(0, 255), color='steelblue', edgecolor='none')
axes[1, 0].set_xlabel('Gray Level Values')
axes[1, 0].set_ylabel('Number of Pixels')
axes[1, 0].set_title('Histogram – Original', fontsize=12)
axes[1, 0].axvline(x=p3,  color='red',   linestyle='--', label=f'3rd pct ({p3:.0f})')
axes[1, 0].axvline(x=p80, color='green', linestyle='--', label=f'80th pct ({p80:.0f})')
axes[1, 0].legend()

# --- Rescaled Image (3rd–80th percentile) ---
axes[0, 1].imshow(img_rescale_3_80, cmap='gray')
axes[0, 1].set_title('Contrast Stretched (3rd–80th %)', fontsize=13, fontweight='bold')
axes[0, 1].axis('off')

axes[1, 1].hist(img_rescale_3_80.flat, bins=256, range=(0, 255), color='mediumseagreen', edgecolor='none')
axes[1, 1].set_xlabel('Gray Level Values')
axes[1, 1].set_ylabel('Number of Pixels')
axes[1, 1].set_title('Histogram – Contrast Stretched (3rd–80th %)', fontsize=12)

plt.suptitle('Assessment Task #1 – Contrast Stretching (3rd–80th Percentiles)', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('assessment_task1_contrast_stretch.png', dpi=150, bbox_inches='tight')
plt.show()
print("Assessment Task #1 completed ✅")
```

## 🟢 Assessment Task #2 – Histogram Equalization

```python
# ─────────────────────────────────────────────────────────────────
# ASSESSMENT TASK #2
# Histogram Equalization using exposure.equalize_hist()
# ─────────────────────────────────────────────────────────────────

img = data.moon()

# Apply histogram equalization
img_eq = exposure.equalize_hist(img)

# Plot results
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# --- Original Image ---
axes[0, 0].imshow(img, cmap='gray')
axes[0, 0].set_title('Original Moon Image', fontsize=13, fontweight='bold')
axes[0, 0].axis('off')

axes[1, 0].hist(img.flat, bins=256, range=(0, 255), color='steelblue', edgecolor='none')
axes[1, 0].set_xlabel('Gray Level Values')
axes[1, 0].set_ylabel('Number of Pixels')
axes[1, 0].set_title('Histogram – Original (Non-Uniform)', fontsize=12)

# --- Equalized Image ---
axes[0, 1].imshow(img_eq, cmap='gray')
axes[0, 1].set_title('Histogram Equalized Image', fontsize=13, fontweight='bold')
axes[0, 1].axis('off')

axes[1, 1].hist(img_eq.flat, bins=256, range=(0, 1), color='coral', edgecolor='none')
axes[1, 1].set_xlabel('Gray Level Values (normalized 0–1)')
axes[1, 1].set_ylabel('Number of Pixels')
axes[1, 1].set_title('Histogram – After Equalization (Flat/Uniform)', fontsize=12)

plt.suptitle('Assessment Task #2 – Histogram Equalization', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('assessment_task2_histogram_equalization.png', dpi=150, bbox_inches='tight')
plt.show()
print("Assessment Task #2 completed ✅")
```

## 🟢 Assessment Task #3 – Histogram Matching (Rocket → Chelsea)

```python
# ─────────────────────────────────────────────────────────────────
# ASSESSMENT TASK #3
# Histogram Matching: Chelsea (source) matched to Rocket (reference)
# ─────────────────────────────────────────────────────────────────

from skimage.exposure import match_histograms

# Load images
reference = img_as_float(data.rocket())   # Reference image
image     = img_as_float(data.chelsea())  # Source image to be matched

# Apply histogram matching
# channel_axis=-1 because both images are RGB (3 channels)
matched = match_histograms(image, reference, channel_axis=-1)

# ── Plot the three images ──
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].imshow(reference)
axes[0].set_title('Reference Image\n(Rocket)', fontsize=13, fontweight='bold')
axes[0].axis('off')

axes[1].imshow(image)
axes[1].set_title('Source Image\n(Chelsea – Before Matching)', fontsize=13, fontweight='bold')
axes[1].axis('off')

axes[2].imshow(matched)
axes[2].set_title('Matched Image\n(Chelsea – After Matching)', fontsize=13, fontweight='bold')
axes[2].axis('off')

plt.suptitle('Assessment Task #3 – Histogram Matching', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('assessment_task3_histogram_matching_images.png', dpi=150, bbox_inches='tight')
plt.show()

# ── Plot Histograms for each channel ──
fig, axes = plt.subplots(3, 3, figsize=(18, 14))
channel_names = ['Red', 'Green', 'Blue']
channel_colors = ['red', 'green', 'blue']

for ch in range(3):
    # Reference histogram
    axes[ch, 0].hist(reference[..., ch].ravel(), bins=256, range=(0, 1),
                     color=channel_colors[ch], alpha=0.7, edgecolor='none')
    axes[ch, 0].set_title(f'Reference – {channel_names[ch]} Channel', fontsize=11)
    axes[ch, 0].set_ylabel('Pixel Count')

    # Source histogram (before)
    axes[ch, 1].hist(image[..., ch].ravel(), bins=256, range=(0, 1),
                     color=channel_colors[ch], alpha=0.7, edgecolor='none')
    axes[ch, 1].set_title(f'Source (Chelsea) – {channel_names[ch]} Channel', fontsize=11)

    # Matched histogram (after)
    axes[ch, 2].hist(matched[..., ch].ravel(), bins=256, range=(0, 1),
                     color=channel_colors[ch], alpha=0.7, edgecolor='none')
    axes[ch, 2].set_title(f'Matched – {channel_names[ch]} Channel', fontsize=11)

# Add column headers
col_labels = ['Reference (Rocket)', 'Source – Before', 'Matched – After']
for ax, col in zip(axes[0], col_labels):
    ax.annotate(col, xy=(0.5, 1.15), xycoords='axes fraction',
                ha='center', fontsize=12, fontweight='bold')

plt.suptitle('Assessment Task #3 – Histogram Matching (Per-Channel Histograms)',
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('assessment_task3_histogram_matching_histograms.png', dpi=150, bbox_inches='tight')
plt.show()

print("Assessment Task #3 completed ✅")
print("\n=== ALL LAB 4 TASKS COMPLETED SUCCESSFULLY ===")
```
