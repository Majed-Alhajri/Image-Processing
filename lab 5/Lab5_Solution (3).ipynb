```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from skimage import data
from PIL import Image

# Create images folder
os.makedirs('images', exist_ok=True)

# Use a built-in sample image from skimage (no internet needed)
parrot = data.coffee()  # Built-in colorful image
img_pil = Image.fromarray(parrot)
img_pil.save('images/Parrot.png')

print('Image created successfully!')
plt.imshow(parrot)
plt.title('Sample Image')
plt.axis('off')
plt.show()
```

## Task #1: Convolve an image with a 7x7 Box Filter
Show the original and smoothed image side by side.

```python
# Read the image
image = cv2.imread('images/Parrot.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply 7x7 Box Filter (normalized)
box_filtered = cv2.boxFilter(image, ddepth=-1, ksize=(7, 7), normalize=True)

# Plot original and smoothed images side by side
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].imshow(image)
ax[0].set_title('Original Image', fontsize=14)
ax[0].axis('off')

ax[1].imshow(box_filtered)
ax[1].set_title('Smoothed Image (7x7 Box Filter)', fontsize=14)
ax[1].axis('off')

plt.tight_layout()
plt.show()
```

## Task #2: Apply 5x5 and 21x21 Gaussian Filters
Show the original image, 5x5 filtered, and 21x21 filtered side by side.

```python
# Read the image
image = cv2.imread('images/Parrot.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply 5x5 Gaussian Filter
gaussian_5x5 = cv2.GaussianBlur(image, (5, 5), sigmaX=0, sigmaY=0)

# Apply 21x21 Gaussian Filter
gaussian_21x21 = cv2.GaussianBlur(image, (21, 21), sigmaX=0, sigmaY=0)

# Plot the three images side by side
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

ax[0].imshow(image)
ax[0].set_title('Original Image', fontsize=14)
ax[0].axis('off')

ax[1].imshow(gaussian_5x5)
ax[1].set_title('Gaussian Filter 5x5', fontsize=14)
ax[1].axis('off')

ax[2].imshow(gaussian_21x21)
ax[2].set_title('Gaussian Filter 21x21', fontsize=14)
ax[2].axis('off')

plt.tight_layout()
plt.show()
```

## Task #3: Sharpen an Image Using Laplacian Filter
Steps:
1. Read the image
2. Convert to float32
3. Apply cv2.Laplacian()
4. Sharpen using: `np.clip(image_float - laplacian, 0, 1)`
5. Plot all three images

```python
# Step 1: Read the image
image = cv2.imread('images/Parrot.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 2: Convert image to float32
image_float = image.astype(np.float32) / 255.0

# Step 3: Apply 3x3 Laplacian filter
laplacian = cv2.Laplacian(image_float, ddepth=cv2.CV_32F, ksize=3)

# Step 4: Sharpen the image
sharpened = np.clip(image_float - laplacian, 0, 1)

# Step 5: Plot the original, Laplacian result, and sharpened images
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

ax[0].imshow(image)
ax[0].set_title('Original Image', fontsize=14)
ax[0].axis('off')

# Normalize Laplacian for display
laplacian_display = np.clip((laplacian - laplacian.min()) / (laplacian.max() - laplacian.min()), 0, 1)
ax[1].imshow(laplacian_display)
ax[1].set_title('Laplacian Filter Result', fontsize=14)
ax[1].axis('off')

ax[2].imshow(sharpened)
ax[2].set_title('Sharpened Image (Laplacian)', fontsize=14)
ax[2].axis('off')

plt.tight_layout()
plt.show()
```
