## Task 1: Geometric Transformations

### Task 1.1: Increase the size of the image (Scaling/Resize)

```python
img = cv2.imread('./images/input.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

original_height, original_width = img.shape[:2]
print(f'Original size: {original_width} x {original_height}')

# Increase size by 2x
scale_factor = 2.0
new_width = int(original_width * scale_factor)
new_height = int(original_height * scale_factor)

img_enlarged = cv2.resize(img, (new_width, new_height), interpolation=cv2.INTER_LINEAR)
img_enlarged_rgb = cv2.cvtColor(img_enlarged, cv2.COLOR_BGR2RGB)

print(f'Enlarged size: {new_width} x {new_height}')

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
axes[0].imshow(img_rgb)
axes[0].set_title(f'Original ({original_width}x{original_height})')
axes[0].axis('off')

axes[1].imshow(img_enlarged_rgb)
axes[1].set_title(f'Enlarged 2x ({new_width}x{new_height})')
axes[1].axis('off')

plt.tight_layout()
plt.savefig('./images/task1_1_enlarged.jpg', bbox_inches='tight')
plt.show()
```

### Task 1.2: Rotate the image by 120 degrees

```python
img = cv2.imread('./images/input.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

h, w = img.shape[:2]
center = (w // 2, h // 2)
angle = 120

rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale=1.0)

# Calculate new bounding dimensions so entire image is visible
cos = np.abs(rotation_matrix[0, 0])
sin = np.abs(rotation_matrix[0, 1])
new_w = int((h * sin) + (w * cos))
new_h = int((h * cos) + (w * sin))

rotation_matrix[0, 2] += (new_w / 2) - center[0]
rotation_matrix[1, 2] += (new_h / 2) - center[1]

img_rotated = cv2.warpAffine(img, rotation_matrix, (new_w, new_h),
                              borderMode=cv2.BORDER_CONSTANT, borderValue=(255, 255, 255))
img_rotated_rgb = cv2.cvtColor(img_rotated, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
axes[0].imshow(img_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(img_rotated_rgb)
axes[1].set_title('Rotated 120 Degrees')
axes[1].axis('off')

plt.tight_layout()
plt.savefig('./images/task1_2_rotated.jpg', bbox_inches='tight')
plt.show()

print(f'Rotation angle: {angle} degrees')
print('Rotation Matrix:\n', rotation_matrix)
```

### Task 1.3: Perform Shear operations over the image

```python
img = cv2.imread('./images/input.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

h, w = img.shape[:2]

# Horizontal Shear
shear_x = 0.3
M_shear_x = np.float32([[1, shear_x, 0], [0, 1, 0]])
new_w_x = int(w + abs(shear_x) * h)
img_shear_x = cv2.warpAffine(img, M_shear_x, (new_w_x, h),
                              borderMode=cv2.BORDER_CONSTANT, borderValue=(255, 255, 255))

# Vertical Shear
shear_y = 0.3
M_shear_y = np.float32([[1, 0, 0], [shear_y, 1, 0]])
new_h_y = int(h + abs(shear_y) * w)
img_shear_y = cv2.warpAffine(img, M_shear_y, (w, new_h_y),
                              borderMode=cv2.BORDER_CONSTANT, borderValue=(255, 255, 255))

img_shear_x_rgb = cv2.cvtColor(img_shear_x, cv2.COLOR_BGR2RGB)
img_shear_y_rgb = cv2.cvtColor(img_shear_y, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
axes[0].imshow(img_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(img_shear_x_rgb)
axes[1].set_title(f'Horizontal Shear (shear_x={shear_x})')
axes[1].axis('off')

axes[2].imshow(img_shear_y_rgb)
axes[2].set_title(f'Vertical Shear (shear_y={shear_y})')
axes[2].axis('off')

plt.tight_layout()
plt.savefig('./images/task1_3_shear.jpg', bbox_inches='tight')
plt.show()

print('Horizontal Shear Matrix:\n', M_shear_x)
print('Vertical Shear Matrix:\n', M_shear_y)
```

## Task 2: Intensity Transformations

### Task 2.1: Negative Transformation

```python
img = cv2.imread('./images/input.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Negative transformation: s = 255 - r
img_negative = 255 - img
img_negative_rgb = cv2.cvtColor(img_negative, cv2.COLOR_BGR2RGB)

cv2.imwrite('./images/task2_1_negative.jpg', img_negative)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
axes[0].imshow(img_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(img_negative_rgb)
axes[1].set_title('Negative Image  (s = 255 - r)')
axes[1].axis('off')

plt.tight_layout()
plt.savefig('./images/task2_1_negative_result.jpg', bbox_inches='tight')
plt.show()

print('Negative Transformation Formula: s = 255 - r')
print('Reverses all intensity values: dark becomes bright and vice versa.')
```

### Task 2.2: Log Transformation

```python
img = cv2.imread('./images/input.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Log transformation: s = c * log(1 + r)
# Use max possible pixel value (255) to compute c safely
# This avoids divide-by-zero when the image has dark/black pixels
c = 255 / np.log(1 + 255)  # fixed denominator = log(256)
print(f'Chosen constant c = {c:.4f}  (based on max possible value 255)')

img_log = np.clip(c * np.log(1 + img.astype(np.float64)), 0, 255).astype(np.uint8)
img_log_rgb = cv2.cvtColor(img_log, cv2.COLOR_BGR2RGB)
cv2.imwrite('./images/task2_2_log.jpg', img_log)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
axes[0].imshow(img_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(img_log_rgb)
axes[1].set_title(f'Log Transformation  s = {c:.2f} * log(1 + r)')
axes[1].axis('off')

plt.tight_layout()
plt.savefig('./images/task2_2_log_result.jpg', bbox_inches='tight')
plt.show()

print('Log Transformation Formula: s = c * log(1 + r)')
print('Brightens dark pixels and compresses bright pixel values.')
```

### Task 2.3: Power Law (Gamma) Transformation — Increase Contrast

```python
img = cv2.imread('./images/input.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Power law: s = c * r^gamma
# gamma > 1 increases contrast in bright areas
gamma = 1.8

img_gamma = np.clip(np.power(img / 255.0, gamma) * 255, 0, 255).astype(np.uint8)
img_gamma_rgb = cv2.cvtColor(img_gamma, cv2.COLOR_BGR2RGB)
cv2.imwrite('./images/task2_3_gamma.jpg', img_gamma)

# gamma < 1 brightens the image
gamma_bright = 0.5
img_gamma_bright = np.clip(np.power(img / 255.0, gamma_bright) * 255, 0, 255).astype(np.uint8)
img_gamma_bright_rgb = cv2.cvtColor(img_gamma_bright, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
axes[0].imshow(img_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(img_gamma_rgb)
axes[1].set_title(f'Power Law  gamma={gamma}  (Increases Contrast)')
axes[1].axis('off')

axes[2].imshow(img_gamma_bright_rgb)
axes[2].set_title(f'Power Law  gamma={gamma_bright}  (Brightens Image)')
axes[2].axis('off')

plt.tight_layout()
plt.savefig('./images/task2_3_gamma_result.jpg', bbox_inches='tight')
plt.show()

print('Power Law (Gamma) Formula: s = r^gamma')
print(f'gamma={gamma} -> Increases contrast in bright regions')
print(f'gamma={gamma_bright} -> Brightens dark regions')
```

## Summary: All Transformations

```python
img = cv2.imread('./images/input.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
h, w = img.shape[:2]

# Geometric
img_resized       = cv2.resize(cv2.resize(img, (w*2, h*2)), (w, h))
M_rot             = cv2.getRotationMatrix2D((w//2, h//2), 120, 1.0)
img_rot           = cv2.warpAffine(img, M_rot, (w, h), borderValue=(255,255,255))
img_shear         = cv2.warpAffine(img, np.float32([[1,0.3,0],[0,1,0]]), (int(w+0.3*h), h), borderValue=(255,255,255))
img_shear_display = cv2.resize(img_shear, (w, h))

# Intensity
img_neg  = 255 - img
c_log    = 255 / np.log(1 + 255)
img_log  = np.clip(c_log * np.log(1 + img.astype(np.float64)), 0, 255).astype(np.uint8)
img_gam  = np.clip(np.power(img/255.0, 1.8) * 255, 0, 255).astype(np.uint8)

images = [
    (img_rgb,                                              'Original'),
    (cv2.cvtColor(img_resized,       cv2.COLOR_BGR2RGB),   'Resize (2x)'),
    (cv2.cvtColor(img_rot,           cv2.COLOR_BGR2RGB),   'Rotate 120 deg'),
    (cv2.cvtColor(img_shear_display, cv2.COLOR_BGR2RGB),   'Shear (x=0.3)'),
    (cv2.cvtColor(img_neg,           cv2.COLOR_BGR2RGB),   'Negative'),
    (cv2.cvtColor(img_log,           cv2.COLOR_BGR2RGB),   f'Log (c={c_log:.1f})'),
    (cv2.cvtColor(img_gam,           cv2.COLOR_BGR2RGB),   'Power Law (gamma=1.8)'),
]

fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()
for i, (image, title) in enumerate(images):
    axes[i].imshow(image)
    axes[i].set_title(title, fontsize=13)
    axes[i].axis('off')
axes[-1].axis('off')

plt.suptitle('Lab 3 - Summary of All Image Transformations', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('./images/summary_all_transformations.jpg', bbox_inches='tight', dpi=150)
plt.show()

print('All transformations completed!')
```
